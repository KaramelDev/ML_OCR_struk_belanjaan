````python
import easyocr
from PIL import Image
import re
import json

reader = easyocr.Reader(['id'], gpu=False)

def OCR_EasyOCR(image_path):
    results = reader.readtext(image_path, detail=0)
    lines = [line.strip().lower() for line in results]
    
    ocr_json = {
        "invoice_number": None,
        "phone": None,
        "alamat": None,
        "email": None,
        "nama_toko": None,
        "tanggal": None,
        "daftar_barang": [],
        "total_belanja": None
    }

    for line in lines:
        if re.search(r"\d{2}/\d{2}/\d{4}", line):
            ocr_json["tanggal"] = re.search(r"\d{2}/\d{2}/\d{4}", line).group()
        if "@" in line:
            ocr_json["email"] = line
        if any(word in line for word in ["jalan", "jl.", "alamat"]):
            ocr_json["alamat"] = line
        if re.search(r"(telp|hp|phone|wa)", line):
            ocr_json["phone"] = re.sub(r"\D", "", line)
        if ocr_json["nama_toko"] is None and len(line) > 4:
            ocr_json["nama_toko"] = line

    for line in lines:
        if "total" in line:
            angka = re.findall(r"\d+[.,]?\d*", line)
            if angka:
                try:
                    ocr_json["total_belanja"] = float(angka[-1].replace(",", "").replace(".", ""))
                except:
                    pass

    for line in lines:
        if any(char.isdigit() for char in line) and any(c.isalpha() for c in line):
            tokens = line.split()
            if len(tokens) >= 2:
                nama = " ".join(tokens[:-1])
                try:
                    harga = float(tokens[-1].replace(",", "").replace(".", ""))
                except:
                    harga = 0.0
                ocr_json["daftar_barang"].append({
                    "nama": nama,
                    "qty": 1,
                    "harga_satuan": harga,
                    "subtotal": harga
                })

    return ocr_json

if __name__ == "__main__":
    image_path = "sample/Receipt NIVEA/ini.jpg"
    hasil = OCR_EasyOCR(image_path)
    print(json.dumps(hasil, indent=2, ensure_ascii=False))
````

````python
import pytesseract
from PIL import Image
import re
import json

# Pastikan path pytesseract sudah sesuai jika belum otomatis
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

def OCR_Pytesseract(image_path):
    img = Image.open(image_path)
    text = pytesseract.image_to_string(img, lang='eng')

    lines = text.split('\n')
    lines = [line.strip().lower() for line in lines if line.strip()]

    ocr_json = {
        "invoice_number": None,
        "phone": None,
        "alamat": None,
        "email": None,
        "nama_toko": None,
        "tanggal": None,
        "daftar_barang": [],
        "total_belanja": None
    }

    # Ekstrak informasi dasar
    for line in lines:
        if re.search(r'\d{2}/\d{2}/\d{4}', line):
            ocr_json["tanggal"] = re.search(r'\d{2}/\d{2}/\d{4}', line).group()
        if '@' in line:
            ocr_json["email"] = line
        if any(x in line for x in ["jalan", "jl.", "alamat"]):
            ocr_json["alamat"] = line
        if re.search(r'(telp|phone|wa|hp)', line):
            number = re.sub(r'\D', '', line)
            if len(number) >= 8:
                ocr_json["phone"] = number
        if ocr_json["nama_toko"] is None and len(line) > 3:
            ocr_json["nama_toko"] = line

    # Total belanja
    for line in lines:
        if 'total' in line:
            angka = re.findall(r'\d+[.,]?\d*', line)
            if angka:
                try:
                    ocr_json["total_belanja"] = float(angka[-1].replace(',', '').replace('.', ''))
                except:
                    pass

    # Deteksi daftar barang (sederhana)
    for line in lines:
        tokens = line.split()
        if len(tokens) >= 2 and any(c.isdigit() for c in tokens[-1]):
            nama = ' '.join(tokens[:-1])
            try:
                harga = float(tokens[-1].replace(',', '').replace('.', ''))
            except:
                harga = 0.0
            ocr_json["daftar_barang"].append({
                "nama": nama,
                "qty": 1,
                "harga_satuan": harga,
                "subtotal": harga
            })

    return ocr_json


# ==== Contoh Penggunaan ====
if __name__ == "__main__":
    image_path = "sample/Receipt NIVEA/ini.jpg"
    hasil = OCR_Pytesseract(image_path)
    print(json.dumps(hasil, indent=2, ensure_ascii=False))
````
